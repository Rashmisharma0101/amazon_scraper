```python
def check_price():
    # connect to website

    url =  'https://in.puma.com/in/en/pd/mens-slim-fit-polo-t-shirt/687093?size=0130&swatch=01&utm_source=BING-DDA&utm_medium=DSP&utm_campaign=DSP_BING_DDA_IN_PMAX_agency_1000067495857508873&msclkid=42cc927bcce51d3feefdfbdab940dbd0'
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36 Edg/145.0.0.0"}
    page = requests.get(url, headers = headers)
    soup1 = BeautifulSoup(page.content, "html.parser")
    soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
    title = soup2.find(id = "pdp-product-title").get_text()
    title = title.strip()
    price = soup2.find("span", {"data-test-id" : "item-sale-price-pdp"}).get_text()
    price = price.strip()
    header = ['title', 'price']

    data = [title, price]

    with open('amazon_web_scrapper.csv', 'a+', newline = '', encoding = 'UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(header)
        writer.writerow(data)
    
```

```python
while (True):
    check_price()
    time.sleep(5)
```
